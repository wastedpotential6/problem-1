```python
import pandas as pd
import sklearn
import numpy as np
```

```python
df_train = pd.read_csv('train 1.csv')
print (df_train)
```

```python
df_test = pd.read_csv('test 1.csv')
print (df_test)
```

```python
zeros_array = np.zeros(133)
df_test_inserted = pd.read_csv('test 1.csv')
df_test_inserted.insert(1,'Onshore/Offshore',zeros_array)
df_test_inserted
```

```python
df_concat = df_train.append(df_test_inserted,ignore_index=True)
df_concat
```

```python
inputs = df_concat.drop('Onshore/Offshore',axis='columns')
target = df_train['Onshore/Offshore']
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_Tectonicregime = LabelEncoder()
le_Hydrocarbontype = LabelEncoder()
le_Reservoirstatus = LabelEncoder()
le_Structuralsetting = LabelEncoder()
le_Period = LabelEncoder()
le_Lithology = LabelEncoder()
```

```python
inputs['Tectonic regime_n'] = le_Tectonicregime.fit_transform(inputs['Tectonic regime'])
inputs['Hydrocarbon type_n'] = le_Hydrocarbontype.fit_transform(inputs['Hydrocarbon type'])
inputs['Reservoir status_n'] = le_Reservoirstatus.fit_transform(inputs['Reservoir status'])
inputs['Structural setting_n'] = le_Structuralsetting.fit_transform(inputs['Structural setting'])
inputs['Period_n'] = le_Period.fit_transform(inputs['Period'])
inputs['Lithology_n'] = le_Lithology.fit_transform(inputs['Lithology'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Tectonic regime','Hydrocarbon type','Reservoir status','Structural setting','Period','Lithology'],axis='columns')
inputs_n
```

```python
inputs_split = np.split(inputs_n, [309], axis=0)
inputs_split[0]
```

```python
inputs_split[1]
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_split[0], target)
```

```python
model.score(inputs_split[0],target)
```

```python
answer = model.predict(inputs_split[1])
answer
```

```python
answer.shape
```

```python
answer_df = pd.DataFrame(data=answer)
answer_df
```
